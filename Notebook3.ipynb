## Final Title - Are AI Generated Datasets Reflective of Real Trends?

## Description

This project is a comparative data analysis presented as a Canva infographic that compares an AI-generated dataset with real world labor market data. Using the AI dataset, I first created a series of graphs illustrating numerical relationships within the data set, relationships between salary and experience level, and job market growth predictions. I then recreated those same visualizations using data from US labor sources to highlight how the AI-produced patterns are different from actual economic trends. The infographic includes detailed explanations of the AI dataset, descriptions of the real world graphs used for comparison, and information on the differences between them, for example, missing correlations in the AI data that are strong in reality, unrealistic similarity in job projections, and inaccurate salary/experience relationships. Overall, the product demonstrates how AI-generated data can appear structured but still fail to reflect the complex and human patterns present in real labor markets.

#### Sources Used in Table on Infographic

* https://research.upjohn.org/cgi/viewcontent.cgi?article=1319&context=reports

This report analyzes how AI exposure across occupations links to U.S. employment projections and potential shifts in future labor demand.

BLS — “Growth trends for selected occupations considered at risk from automation” (2022, Monthly Labor Review)
* https://www.bls.gov/opub/mlr/2022/article/growth-trends-for-selected-occupations-considered-at-risk-from-automation.htm
